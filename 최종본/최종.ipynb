```python
%%writefile d:\myPyCode\팀프로젝트\netflix_streaming_API_module.py

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import csv
from matplotlib import font_manager, rc

df = pd.read_csv("All_Streaming_Shows.csv",encoding="latin1")
select = pd.DataFrame(df)
select = select.fillna(method='ffill')

def API_netflix_data_init():
    df = pd.read_csv("All_Streaming_Shows.csv",encoding="latin1")
    select = pd.DataFrame(df)
    select = select.fillna(method='ffill')
    return select


# 날짜별순으로 값 원하는 값 뽑아내기 (type과 key값 )
def API_netflix_data_findint(select_col,Name,wants_type_standard) :
    #select.sort_values(by = ['release_year','date_added']) 날짜별로 구분하기
    selection = select[select[select_col] == Name].sort_values(by = wants_type_standard)
    #select[select[select_col] == Name] Name과 같은 select key값을 뽑아낸다.
    return selection

# 날짜순으로 타입안에 원하는 문자포함되어 있는 행 뽑아내기
def API_netflix_data_findword(netflix_framedata,position,Name,wants_type_standard) :
    find_line = netflix_framedata.loc[netflix_framedata[position].str.contains(Name)].sort_values(by= 
    wants_type_standard)
    return find_line

# 중복 없이 타입 뽑아내기
def API_remove_overlap(netflix_data,netflix_want_name):
    list_module = []
    netflix_data_find = set(list(netflix_data[netflix_want_name]))
    for i in netflix_data_find :
        list_module.append(i)
    return list_module

# 데이터프레임에서 반복 갯수세기
def API_netflix_count(netflix_dataframe,collect_type) :
    collect_lists = netflix_dataframe[collect_type].value_counts()
    return collect_lists


# 문자로 되어있는 데이터 프레임 원하는 keys값의 values 개수세기
def API_netflix_data_sort(netflix_data,col_keys,col_vaules,sort_type1,sort_type2) :
    
    #원하는장르포함 모든 행 뽑기
    listed_in_pullout = API_netflix_data_findword(netflix_data,col_keys, col_vaules,sort_type1)
    # 시대별 기준으로 나누기
    listed_in_count = API_netflix_count(listed_in_pullout,sort_type2) 
    # 아이템으로 정렬하면 key값을 정렬가능 내림차순
    listed_in_count = dict(sorted(listed_in_count.items(),reverse=True))
    
    return listed_in_count


def graph_step5(select,years_listed_in,Genre_lists,streamings,colors):
    streaming_data = pd.DataFrame()
    # dataframe 초기 설정
    streaming_data = streaming_data.reindex(index=streamings)
    font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
    rc('font', family=font_name)
    num=0
    fig = plt.figure(figsize=(9,9)) 

    for years_list in years_listed_in:
        New_select = API_netflix_data_findint('Year Released',years_list,"Streaming Platform")
        streaming_data[Genre_lists] = 0
        num=num+1

        for Genre_list in Genre_lists :
            Genres_count= {}
            listed_in_count = API_netflix_data_sort(New_select,'Genre',Genre_list,'Year Released','Streaming Platform')
            
            for streaming in streamings :
                Genre_countcount=0

                for list_in_count in listed_in_count.keys() :
                    if streaming in list_in_count :
                        Genre_countcount = Genre_countcount + int(listed_in_count[list_in_count])
                    
                    Genres_count[streaming]=Genre_countcount
                    streaming_data.loc[streaming,[Genre_list]]=Genres_count[streaming]
            
            
        ax = streaming_data.plot(kind='bar',stacked=True,fontsize=5,color=colors)
        plt.xticks(rotation =30,fontsize=8)
        plt.legend(loc='best',fontsize=8,shadow = True)
        plt.title(str(years_list)+'년 장르별 스트리밍 비교',fontsize=10)
        #https://kongdols-room.tistory.com/86
```

```python
%%writefile d:\myPyCode\팀프로젝트\netflix_frame_module.py


import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import csv
import seaborn as sns
import json


df = pd.read_csv("netflix_titles.csv",encoding="latin1")
select = pd.DataFrame(df,columns = ['show_id', 'type', 'title','country', 'date_added','release_year', 'rating',
'duration','listed_in', 'description'])

# 빈 값 처리
select = select.fillna(method='ffill')


# 초기 데이터프레임 뽑아내기
def netflix_data_init() :
    df = pd.read_csv("netflix_titles.csv",encoding="latin1")
    select = pd.DataFrame(df,columns = ['show_id', 'type', 'title','country', 'date_added','release_year',
    'rating','duration','listed_in', 'description'])
    return select


# DataFrame 열 기준으로 똑같은 타입 values값 뽑아내기
def netflix_data_findint(netflix_data,select_col,Name) :
    selection = netflix_data[netflix_data[select_col] == Name].sort_values(by = ['release_year','date_added'])
    return selection
    #select.sort_values(by = ['release_year','date_added']) 날짜별로 구분하기
    #select[select[select_col] == Name] Name과 같은 select key값을 뽑아낸다.


# DataFrame 열기준으로 포함 문자 values값 뽑아내기
def netflix_data_findword(netflix_framedata,position,Name) :
    find_line = netflix_framedata.loc[netflix_framedata[position].str.contains(Name)].sort_values(by= 
    ['release_year','date_added'])
    return find_line


# DataFrame 열기준으로 중복 제거 values값 뽑아내기
def remove_overlap(netflix_data,netflix_want_name):
    list_module = []
    netflix_data_find = set(list(netflix_data[netflix_want_name]))
    for i in netflix_data_find :
        list_module.append(i)
    return list_module


# DataFrame 열기준으로 같은 values값 개수세기
def netflix_count(netflix_dataframe,collect_type) :
    collect_lists = netflix_dataframe[collect_type].value_counts()
    return collect_lists
    

# 문자로 되어있는 데이터 프레임 원하는 keys값의 values 개수세기
def netflix_data_sort(netflix_data,col_keys,col_vaules,sort_type) :
    
    #원하는장르포함 모든 행 뽑기
    listed_in_pullout = netflix_data_findword(netflix_data,col_keys, col_vaules)
    # 시대별 기준으로 나누기
    listed_in_count = netflix_count(listed_in_pullout,sort_type) 
    # 아이템으로 정렬하면 key값을 정렬가능 내림차순
    listed_in_count = dict(sorted(listed_in_count.items(),reverse=True))
    
    return listed_in_count


# 이전 데이터프레임통해서 새로운 데이터 프레임을 만든다.
def netflix_make_new_dataframe(new_datafame,old_dataframe,new_dataframe_index,old_col):
    for index_elements in new_dataframe_index :
            if index_elements in list(old_dataframe.keys()) :
                new_datafame.loc[index_elements,[old_col]]=old_dataframe[index_elements]

    return new_datafame



# 데이터 정렬하기
def netflix_data_years_clear(netflix_data,wants_col_Name,wants_index_Name,neflix_col_Name):
    
    # 시대별 영화장르 뽑아내기 새로운 데이터프레임 생성
    df = pd.DataFrame( )
    df = df.reindex(index = wants_index_Name)
    df[wants_col_Name] = 0

    for movie_elements in wants_col_Name :
        # 개수를 세고 정렬하기
        listed_in_count = netflix_data_sort(netflix_data,neflix_col_Name,movie_elements,'release_year')
        # dataframe 초기 설정
        df = df.reindex(index=wants_index_Name)
        # 새로운 데이터프레임 개수별 index,col 지정
        netflix_make_new_dataframe(df,listed_in_count,wants_index_Name,movie_elements)
    return df

# 새로운 이름의 열값에 합쳐진 데이터 프레임 만들기
def new_dataframe_rating(netflix_dataFrame,new_dataframe_col):
    new_dataFrame_netflix = pd.DataFrame()
    new_dataFrame_netflix = new_dataFrame_netflix.reindex(index = list(netflix_dataFrame.index))
    new_dataFrame_netflix[list(new_dataframe_col)]=0
    for New_col,New_values in new_dataframe_col.items():
        for New_value in New_values :
            new_dataFrame_netflix[New_col] = new_dataFrame_netflix[New_col] + netflix_dataFrame[New_value]

    return new_dataFrame_netflix


# 나라를 기준으로 데이터 추출
def graph_step1(netflix_data,movie_listed_in,years_country_listed_in,country_listed_in,colors):
    # a=1
    fig = plt.figure(figsize=(9,9)) 
    # 나라별 데이터 뽑기
    Num=1
    for netflix_country_lists in country_listed_in :
        netflix_country_data =  netflix_data_findword(netflix_data,'country',netflix_country_lists)
        netflix_country_dataFrame = netflix_data_years_clear(netflix_country_data,movie_listed_in,years_country_listed_in,'listed_in')

        netflix_country_dataFrame.plot(kind='bar',fontsize=13,title= netflix_country_lists,color=colors)
        plt.xticks(rotation =40,fontsize=13)
        plt.legend(loc='best',fontsize=8,shadow = True)
        plt.show()

# 나라별 시청 등급 갯수세기
def graph_step3(netflix_data,years_age_listed_in,country_listed_in,TV_list,colors):

    fig = plt.figure(figsize=(9,9)) 
    
    # 나이별 데이터 뽑기
    for netflix_country_lists in country_listed_in :
        netflix_country_data =  netflix_data_findword(netflix_data,'country',netflix_country_lists)
        # 연령 시청 종류 조사
        netflix_age_list = remove_overlap(netflix_data,'rating')
        # 연령별 데이터 뽑기
        netflix_age_dataFrame = netflix_data_years_clear(netflix_country_data, netflix_age_list, years_age_listed_in,'rating')
        new_dataFrame_netflix = new_dataframe_rating(netflix_age_dataFrame,TV_list)
       
        new_dataFrame_netflix.plot(kind='bar',fontsize=13,bottom=0,title= netflix_country_lists,color=colors)
        plt.xticks(rotation =40,fontsize=13)
        plt.legend(loc='best',fontsize=7,shadow = True)
        plt.show()

#  데이터 갯수 세기
def netlfix_count_rating(data):
    num=0
    for w in data :
        num += w
    return num

# step 4 나라 하나 당 연도별 그래프 뽑기 
# 그래프 속 내용 : 장르별, 나이별 :  count
def graph_step4_pre(year,nation,netflix_ages,movie_listed_in,TV_list):
    # 년도별 뽑는다.
    select_years = netflix_data_findint(select,'release_year',year)
    # 년도별로 뽑은걸 나라별로 뽑는다.
    select_year = netflix_data_findword(select_years,'country',nation)
    # 비어있는 데이터프레임 생성
    df = pd.DataFrame()
    # 행값의 네임 설정 (netflix_age)
    df = df.reindex(index=netflix_ages)
    

    #데이터 프레임 설정 장르별,나이별 count
    for netflix_age in netflix_ages :
        find_age_year = netflix_data_findword(select_year,'rating',netflix_age)
        
        for i in movie_listed_in :
            select_movie_year = netflix_data_findword(find_age_year,'listed_in',i)
            data = netflix_count(select_movie_year,'listed_in').values
            num = netlfix_count_rating(data)
            df.loc[netflix_age,[i]] = num

    df = df.transpose()
    new_dataFrame_netflix = new_dataframe_rating(df,TV_list)

    plt.figure(figsize=(5,5))
    sns.set(font_scale=2)
    sns.heatmap(new_dataFrame_netflix,annot=True, fmt='0.0f',annot_kws={"size": 8}, cmap='Oranges')

    plt.xticks(rotation =20,fontsize=8)
    plt.yticks(rotation =0,fontsize=8)
    plt.title(str(year)+' - '+nation,fontsize=13)
    plt.show()
    # https://analytics4everything.tistory.com/66 참고 

# 스트리밍 별 장르 갯수 비교
def graph_step4(country_listed_in,movie_listed_in,TV_list):
    netflix_age = remove_overlap(select,'rating')
    for country_list in country_listed_in:
        for year_num in range(2017,2020):
            graph_step4_pre(year_num,country_list,netflix_age,movie_listed_in,TV_list)
    

```
